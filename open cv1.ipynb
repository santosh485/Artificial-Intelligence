```python
#blur
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('f.jpg')
image = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(image,(3,3),0) #kornel
plt.imshow(blur,cmap='gray')
plt.show()
```

```python
#noise high frequency
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('f.jpg')
image = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
blur = cv2.bilateralFilter(image,3,0,0)#filter bilateral medical
plt.imshow(blur,cmap='gray')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img = cv2.imread('f.jpg')
image = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#Nobel filter to remove all noise
snobel= 1/7*np.array([[-1,1,1],
                      [1,5,-2],
                      [1,0,1]]) #normalise
blur=cv2.filter2D(image,-1,snobel)
plt.imshow(blur,cmap='gray')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img = cv2.imread('f.jpg')
image = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#Nobel filter to remove all noise
snobel= 1/24*np.array([[-1,1,1],
                      [1,5,5],
                      [1,5,5]]) #normalise
blur=cv2.filter2D(image,-1,snobel)
plt.imshow(blur,cmap='gray')
plt.show()
```
